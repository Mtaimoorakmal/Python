```python
import json
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Load credentials from JSON file
with open('credentials.json') as f:
    credentials = json.load(f)

username = credentials['username']
password = credentials['password']

# Setup Chrome options
options = Options()
options.add_experimental_option('detach', True)

# Initialize the Chrome driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
driver.maximize_window()

# Open LinkedIn login page
driver.get("https://www.linkedin.com/login")

# Define the wait object
wait = WebDriverWait(driver, 20)

# Wait until the username/email field is present
username_field = wait.until(EC.presence_of_element_located((By.ID, "username")))
username_field.send_keys(username)
print("Username entered.")

# Wait until the password field is present
password_field = wait.until(EC.presence_of_element_located((By.ID, "password")))
password_field.send_keys(password)
print("Password entered.")

# Wait until the login button is clickable and click it
login_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[@type='submit']")))
login_button.click()
print("Login button clicked.")

# Commented out the OTP section
# Sleep time to allow manual entry of OTP
# print("Please enter your OTP within 60 seconds.")
# time.sleep(60)  # Adjust this time as needed to allow for OTP entry

# Verify if the window is still open before proceeding
try:
    driver.current_window_handle
except:
    print("The browser window was closed. Exiting script.")
    exit()

# Wait until the profile icon is present after login
try:
    profile_icon = wait.until(EC.presence_of_element_located((By.XPATH, "//img[contains(@class, 'global-nav__me-photo')]")))
    print("Profile icon located.")
    
    # Click the profile icon to open the dropdown menu
    profile_icon.click()
    print("Profile icon clicked.")

    # Wait until the logout button is present in the dropdown menu and click it
    logout_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//a[@href='/m/logout/']")))
    logout_button.click()
    print("Logout button clicked.")

    # Optional: Add a delay to ensure the logout process completes
    wait.until(EC.presence_of_element_located((By.ID, "login-submit")))
    print("Logged out successfully.")

except Exception as e:
    print("An error occurred:", e)
finally:
    # Close the driver
    driver.quit()
```
