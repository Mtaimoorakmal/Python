```python
import random 
color = ['red' , 'green' , "yellow"]
alien_color = random.choice(color)
print(f'It is "{alien_color}"')
if alien_color =='green':
    print("You earned 5 points")
elif alien_color =='yellow':
    print("You earned 10 points")
else:
    print("You earned 15 points")
```
